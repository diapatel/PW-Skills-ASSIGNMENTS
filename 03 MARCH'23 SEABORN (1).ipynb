<b>QUESTION 1:</b>Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.<br>
<b>SOLUTION : </b>
* Scatter plot - It is used to identify relatioship between 2 variables or set of values.<br>
* Distribution plot - It shows the distribution of data by plotting data as data intervals and the frequencies of values falling into those data intervals.<br>
* Relational plot- When relation between multiple factors have to be identied, this plot is used.<br>
* Categorical plot - It is used to visualise the distribution of categorical data.<br>
* Joint plot - it is a combination of scatter plot and histogram. it displays the scatterplot for 2 numerical set of values and also creates individual histograms for each of them, which shows their distribution.<br>

<b>QUESTION 2:</b> Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x ="timepoint" and y = "signal" for different events and regions.
Note: timepoint, signal, event, and region are columns in the fmri dataset.<br>
<b>SOLUTION : </b><br>

```python
fmri = sns.load_dataset('fmri')
```

```python
sns.lineplot(data = fmri, x='timepoint', y='signal', hue='region')
```

```python
sns.lineplot(data=fmri, x='timepoint', y='signal', hue='event')
```

<b>QUESTION 4:</b>Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue

<b>QUESTION 6:</b>Use the "flights" dataset from seaborn to plot a heatmap.<br>
<b>SOLUTION : </b><br>

```python
flights = sns.load_dataset('flights')
flights = flights.pivot('month', 'year', 'passengers')
```

```python
sns.heatmap(data=flights, cmap='Spectral', annot=True, fmt='d', annot_kws={'size':10})
```
