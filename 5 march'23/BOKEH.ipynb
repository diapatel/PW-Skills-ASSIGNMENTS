QUESTION 1

```python
from bokeh.plotting import figure, show
import numpy as np
from bokeh.io import output_notebook

#creating 2 random arrays
x = np.random.normal(size=100)
y = np.random.normal(size=100)

p = figure(title='scatter plot', x_axis_label ='X-axis', y_axis_label='Y-axis')
p.scatter(x,y)
output_notebook()
show(p)
```

QUESTION 2<Br>
Glyphs in bokeh are graphical markers that can be added to a plot to display data points. 
* glyphs have various customizable properties like shape, tranparency and color
* Bokeh offers various glyph types like circles , squares, triangles, etc.

```python
x = [1,2,3,4,5]
y =[2,3,4,5,6]
z = [1,2,3,5,6]

p = figure(title='Demonstrating bokeh glyphs')

p.circle(x,y)
p.line(x,z, color='red')
output_notebook()
show(p)
```

QUESTION 3<br>
we can use the various parameters provided by the bokeh library to customize the labels, legend and plots as shown in the code cell below.

```python
from bokeh.models import Legend

x = [1,2,3,4,5]
y = [6,7,8,1,8]
z = [1,2,3,4,6]

p = figure(title='Sample plot', x_axis_label='X-axis', y_axis_label='Y-axis')
p.line(x,y, legend_label='sample_plot1')
p.line(x,z,legend_label='sample_plot2', color='red')

legend = Legend(items=[('sample_plot',[p.line(x,y)])])
#p.add_layout(legend,'right')
output_notebook()
show(p)
```

QUESTION 4<Br>
Bokeh server is a Python process that runs a Bokeh application, allowing it to be served to multiple clients over a network. 
* The Bokeh server provides a web interface that allows clients to interact with the Bokeh application by sending events back to the server. 
* The server then updates the plot in response to these events and sends the updated plot back to the client, resulting in a dynamic and interactive visualization that can be updated in real-time.



QUESTION 5<Br>
we can embed a bokeh plot in a webpage using flask or django by first creating a bokeh plot and then rendering it in a HTML template.

```python
from bokeh.plotting import figure, output_file, show

# Generate some data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a Bokeh figure
fig = figure(title='Bokeh Plot', x_axis_label='x', y_axis_label='y')
fig.line(x, y)

# Save the plot to a file
output_file('bokeh_plot.html')

# Show the plot in a web browser
show(fig)
```

```python
from flask import Flask, render_template
from bokeh.embed import file_html
from bokeh.resources import CDN

# Create a Flask app
app = Flask(__name__)

# Define a route for the web page
@app.route('/')
def index():
    # Generate the Bokeh plot
    x = [1, 2, 3, 4, 5]
    y = [6, 7, 2, 4, 5]
    fig = figure(title='Bokeh Plot', x_axis_label='x', y_axis_label='y')
    fig.line(x, y)

    # Embed the Bokeh plot in the HTML template
    plot_html = file_html(fig, CDN, "Bokeh Plot")
    return render_template('index.html', plot_html=plot_html)

# Run the Flask app
if __name__ == '__main__':
    app.run()





```
