* calculating likelihood of the features for each class
P(X1=3 | A) = 4/13 = 0.3077<br>
P(X2=4 | A) = 3/13 = 0.2308<br>
P(X1=3 | B) = 1/6 = 0.1667<br>
P(X2=4 | B) = 1/6 = 0.1667<br>

* calculating posterior priorities for each class
P(A | X1=3, X2=4) = P(X1=3 | A) * P(X2=4 | A) * P(A) / P(X1=3, X2=4)<br>
P(B | X1=3, X2=4) = P(X1=3 | B) * P(X2=4 | B) * P(B) / P(X1=3, X2=4)<br>

* eliminating the common denominators
P(A | X1=3, X2=4) = 0.3077 * 0.2308 * 0.5385 / P(X1=3, X2=4)<br>
P(B | X1=3, X2=4) = 0.1667 * 0.1667 * 0.4615 / P(X1=3, X2=4)<br>

P(X1=3, X2=4) = P(X1=3 | A) * P(X2=4 | A) * P(A) + P(X1=3 | B) * P(X2=4 | B) * P(B)<br>
P(X1=3, X2=4) = 0.3077 * 0.2308 * 0.5385 + 0.1667 * 0.1667 * 0.4615<br>
P(X1=3, X2=4) = 0.0429<br>


P(A | X1=3, X2=4) = 0.3077 * 0.2308 * 0.5385 / 0.0429 = 0.9091<br>
